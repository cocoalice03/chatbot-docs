```python
import numpy as np
import matplotlib.pyplot as plt

#Activité 1
def cout_global(cout_fixe, cout_unitaire, nb_produits):
    return cout_fixe + cout_unitaire * nb_produits

#Paramètres de base
cout_fixe_base = 1000
cout_unitaire_base = 50
nb_produits_max = 100

#1 Variation du coût fixe
cout_fixes = [500, 1000, 1500]
for cf in cout_fixes:
    nb_produits = np.arange(0, nb_produits_max + 1)
    couts = cout_global(cf, cout_unitaire_base, nb_produits)
    plt.plot(nb_produits, couts, label=f'Coût fixe = {cf}')

plt.title("Variation du coût fixe")
plt.xlabel("Nombre de produits")
plt.ylabel("Coût global")
plt.legend()
plt.grid(True)
plt.show()

#2 Variation du coût unitaire
cout_unitaires = [30, 50, 70]
for cu in cout_unitaires:
    nb_produits = np.arange(0, nb_produits_max + 1)
    couts = cout_global(cout_fixe_base, cu, nb_produits)
    plt.plot(nb_produits, couts, label=f'Coût unitaire = {cu}')

plt.title("Variation du coût unitaire")
plt.xlabel("Nombre de produits")
plt.ylabel("Coût global")
plt.legend()
plt.grid(True)
plt.show()

#3 Variation du nombre de produits
nb_produits_max_list = [50, 100, 150]
for npm in nb_produits_max_list:
    nb_produits = np.arange(0, npm + 1)
    couts = cout_global(cout_fixe_base, cout_unitaire_base, nb_produits)
    plt.plot(nb_produits, couts, label=f'Nb produits max = {npm}')

plt.title("Variation du nombre de produits")
plt.xlabel("Nombre de produits")
plt.ylabel("Coût global")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Activité 4
import numpy as np

# Matrice des coefficients
A = np.array([
    [1, 2, -1, 1, 1],
    [2, -1, 3, -1, 4],
    [-1, 3, 2, 1, -1],
    [3, -1, 1, 2, 3],
    [2, 1, -1, -3, 1]
])

# Vecteur des termes constants
Y = np.array([8, 15, 7, 20, 5])


# Calcul du déterminant de A
det_A = np.linalg.det(A)

if det_A != 0:
    print("Le système possède une solution unique.")
else:
    print("Le système n'a pas de solution unique (soit aucune solution, soit une infinité de solutions).")


# Résolution du système
try:
    X = np.linalg.solve(A, Y)
    print("La solution du système est :")
    print(f"x = {X[0]}")
    print(f"y = {X[1]}")
    print(f"z = {X[2]}")
    print(f"t = {X[3]}")
    print(f"h = {X[4]}")
except np.linalg.LinAlgError:
    print("Le système n'a pas de solution unique.")

    # Vérification si les solutions sont entières
    if np.allclose(X, np.round(X)):
        print("Les solutions sont des entiers.")
    else:
        print("Les solutions ne sont pas des entiers.")
except np.linalg.LinAlgError:
    print("Le système n'a pas de solution unique.")
```

```python
#Activite 1
import math
import matplotlib.pyplot as plt
# Paramètres
h = 0.5
f0 = 1

# Méthode d'Euler itérative
f_euler = [f0]
t_euler = [0]
for i in range(3): # Calcul de f1, f2, f3
    f_suivant = f_euler[-1] + h * f_euler[-1] # f_{n+1} = f_n + h*f_n
    f_euler.append(f_suivant)
    t_euler.append(t_euler[-1] + h)

# Valeurs analytiques correspondantes
f_analytique_euler = [math.exp(t) for t in t_euler]

# Affichage des résultats
print("Méthode d'Euler (h=0.5)")
print("-----------------------")
for i in range(1, 4):
    print(f"f{i} (Euler) ≈ {f_euler[i]:.3f}, f({t_euler[i]:.1f}) (Analytique) = {f_analytique_euler[i]:.3f}")

# Graphique comparatif simplifié
plt.figure(figsize=(8, 5))
plt.plot(t_euler, f_analytique_euler, label="Analytique (e^t)", linestyle='-')
plt.plot(t_euler, f_euler, label="Euler (h=0.5)", marker='o', linestyle='--')
plt.xlabel("t")
plt.ylabel("f(t)")
plt.title("Comparaison Euler vs Analytique (équation f'=f)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Activité 2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def euler_method(h, t_final):
    # Étape 1: Initialisation
    N = int(t_final/h)
    tn = [i*h for i in range(N+1)]

    # Étape 2: Génération de la suite d'Euler
    fn = [0]*(N+1)
    fn[0] = 1  # condition initiale

    for n in range(N):
        fn[n+1] = fn[n] + h*fn[n]

    return tn, fn

def solution_exacte(t):
    return np.exp(t)

def calcul_mse(fn, tn):
    # Étape 3: Calcul des erreurs au carré
    exact = [solution_exacte(t) for t in tn]
    erreurs = [(f - e)**2 for f, e in zip(fn, exact)]

    # Étape 4: Calcul de la moyenne (MSE)
    mse = sum(erreurs)/len(erreurs)
    return mse, exact, erreurs

# Paramètres pour les deux valeurs de h
h_values = [0.1, 0.5]
t_final = 2

# Création du graphique
plt.figure(figsize=(10, 6))

# Solution exacte (courbe continue)
t_exact = np.linspace(0, t_final, 200)
plt.plot(t_exact, solution_exacte(t_exact), 'r-', label='Solution exacte', linewidth=2)

# Solutions d'Euler pour différentes valeurs de h
colors = ['b--', 'g--']
markers = ['o', 's']

# Tableau comparatif pour chaque h
for h, color, marker in zip(h_values, colors, markers):
    print(f"\nAnalyse pour h = {h}:")
    print("-" * 50)

    # Calcul des solutions
    tn, fn = euler_method(h, t_final)
    mse, exact, erreurs = calcul_mse(fn, tn)

    # Ajout au graphique
    plt.plot(tn, fn, color, marker=marker, label=f'Euler h={h}', markersize=8)

    # Création du tableau comparatif
    df = pd.DataFrame({
        'tn': tn,
        'Euler (fn)': fn,
        'Exacte (e^tn)': exact,
        'Erreur²': erreurs
    })

    # Affichage des résultats
    print("\nTableau comparatif :")
    print(df.round(6))
    print(f"\nMSE = {mse:.6f}")

# Finalisation du graphique
plt.title("Comparaison des solutions : Euler vs Exacte")
plt.xlabel("t")
plt.ylabel("f(t)")
plt.grid(True)
plt.legend()
plt.show()

# Comparaison finale des MSE
print("\nRésumé de l'impact du pas h sur la MSE:")
print("-" * 50)
for h in h_values:
    tn, fn = euler_method(h, t_final)
    mse, _, _ = calcul_mse(fn, tn)
    print(f"Pour h = {h}: MSE = {mse:.6f}")
```
